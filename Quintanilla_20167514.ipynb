# **Funciones ⭐**

```python
#función con el subconjunto "ranking_años_educacion"
def primeros_10_puestos_con_años_educacion (data,columna,n=10):
  año_educa_ordenado = ranking_años_educacion.sort_values(by=columna,ascending=True)
  primeros_puestos = año_educa_ordenado.head(n)
  print(primeros_puestos)
```

```python
primeros_10_puestos_con_años_educacion(ranking_años_educacion,'Ranking años de educación',10)
```

```python
def ultimos_10_puestos_secundaria_completa(data,columna,n=3):
  secu_completa_ordenada_inversa = ranking_secundaria_completa.sort_values(by='Ranking Educación secundaria completa',ascending=False)
  ultimos_puestos = secu_completa_ordenada_inversa.head(3)
  print(ultimos_puestos)
```

```python
ultimos_10_puestos_secundaria_completa(ranking_secundaria_completa,'Ranking Educación secundaria completa',3)
```
